```python
import matplotlib.pyplot as plt

# Data
gender = [1, 0, 1, 1, 0, 1, 1, 0, 1, 1]
ipk = [3.6, 3.5, 3.8, 2.9, 3.1, 3.4, 2.7, 3.9, 3.3, 3.2]

# Plot
plt.figure(figsize=(10, 4))


# Graduation status 1
ipk_lulus = [ipk[i] for i in range(len(gender)) if gender[i] == 1]
# Graduation status 0
ipk_tidak_lulus = [ipk[i] for i in range(len(gender)) if gender[i] == 0]

plt.scatter([1]*len(ipk_lulus), ipk_lulus, color='green', label='Lulus (1)')
plt.scatter([0]*len(ipk_tidak_lulus), ipk_tidak_lulus, color='black', label='Tidak Lulus (0)')

# Labels and title
plt.xlabel('Status Kelulusan')
plt.ylabel('IPK')
plt.title('Hubungan antara gender dan IPK')
plt.legend()
plt.xticks([0, 1], ['Tidak Lulus', 'Lulus'])
plt.grid(True)
plt.plot( ipk, marker='*', color ='b', linestyle='-')

# Show plot
plt.show()
```

```python
# contoh 4

import seaborn as sns 
import matplotlib.pyplot as plt

# Data contoh
usia = [21,22,20,23,24,21,22,20,23,21]
ipk = [3.9,3.4,3.1,3.8,3.5,3.0,3.7,3.6,3.2,2.9]

# scatter plot #1
plt.figure(figsize=(7,4))
sns.scatterplot(x=usia, y=ipk)
plt.title('Hubungan Antara Usia dan Ipk #1')
plt.xlabel('Usia')
plt.ylabel('Ipk')
plt.grid(True)
plt.show()

# scatter plot #2
plt.figure(figsize=(7,4))
sns.regplot(x=usia, y=ipk)
plt.title('Hubungan Antara Usia dan Ipk #2')
plt.xlabel('Usia')
plt.ylabel('Ipk')
plt.grid(True)
plt.show()
```
